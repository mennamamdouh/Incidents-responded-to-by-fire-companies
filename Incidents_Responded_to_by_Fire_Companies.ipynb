# Question 1

```python
# Show the occurences of each type and order them in descending order
incident_types = incidents_dataset.groupby('INCIDENT_TYPE_DESC').size().sort_values(ascending=False)
incident_types
```

```python
# We see that the most common type is '300 - Rescue, EMS incident, other'
print("The most common type of incident is: {}\nwith number of occurences = {}"
.format(incident_types.index.to_list()[0], incident_types.values[0]))
```

```python
# The proportion of the most common type
print("The propotion of '300 - Rescue, EMS incident, other' is: {}%".format(incident_types.values[0]/incident_types.sum()*100))
```

# Question 1 - Bar chart

```python
# Extract the code of each incident type
types = list(incident_types.index)
codes = []
for type in types:
  temp = type.split(' ')
  code = temp[0]
  codes.append(code)
```

```python
# A bar chart shows the number of incidents for each type
plt.figure(figsize=(150, 30))
plt.title("Types of incidents and its corresponding number of occurences", fontsize=75, fontweight='bold')
plt.xlabel("Type code", fontsize=75)
plt.ylabel("# of occurences", fontsize=75)
plt.xticks(size = 35)
plt.yticks(size = 50)
sns.set(style="whitegrid")
graph = sns.barplot(x=codes, y=incident_types)
graph.set_yscale("log")
graph.tick_params(axis='x', rotation=90)
```

# Question 2

```python
# The ratio of the average number of units of incident
# '111 - Building fire' and incident '651 - Smoke scare, odor of smoke'

# First, select the 2 columns with the INCIDENT_TYPE_DESC and the UNITS_ONSCENE
types_and_units = incidents_dataset.loc[:, ['INCIDENT_TYPE_DESC', 'UNITS_ONSCENE']]

# Second, retreive only the INCIDENT_TYPE_DESC '111 - Building fire' & '651 - Smoke scare, odor of smoke'
types_and_units = types_and_units[incidents_dataset['INCIDENT_TYPE_DESC'].isin(['111 - Building fire', '651 - Smoke scare, odor of smoke'])]

# Thirs, groupby the incident type and get the average for each type
averages = types_and_units.groupby('INCIDENT_TYPE_DESC').mean().values
# Forth, get the ratio between the 2 averages
ratio = averages[0]/averages[1]
```

```python
# Printing the ratio of the averages
print("The ratio of the average number of units of incident 111 and incident 651 = {}".format(float(ratio)))
```

# Question 4

```python
# Get the data for '111 - Building fire' incidents which includes the INCIDENT_TYPE_DESC, INCIDENT_DATE_TIME and ARRIVAL_DATE_TIME columns
building_fire = incidents_dataset[incidents_dataset['INCIDENT_TYPE_DESC'] == '111 - Building fire'].loc[:, ['INCIDENT_TYPE_DESC', 'INCIDENT_DATE_TIME', 'ARRIVAL_DATE_TIME']]

# Let's check the dtype of the columns: INCIDENT_DATE_TIME and ARRIVAL_DATE_TIME
print("The data type of both columns is: {}, {}".format(building_fire.INCIDENT_DATE_TIME.dtype, building_fire.ARRIVAL_DATE_TIME.dtype))

# So, we need to convert them into datetime types to be able to extract the difference in minutes
```

```python
# Importing the module we need
from datetime import datetime
```

```python
# Convert the 2 columns into datetime format
building_fire['INCIDENT_DATE_TIME'] = pd.to_datetime(building_fire['INCIDENT_DATE_TIME'], utc=True)
building_fire['ARRIVAL_DATE_TIME'] = pd.to_datetime(building_fire['ARRIVAL_DATE_TIME'], utc=True)

# Check the data type
print("Now, the data type of both columns is: {}, {}".format(building_fire.INCIDENT_DATE_TIME.dtype, building_fire.ARRIVAL_DATE_TIME.dtype))
```

```python
# Get the difference in minutes between the 2 times
building_fire['TIME_DIFFERENCE'] = building_fire['ARRIVAL_DATE_TIME'] - building_fire['INCIDENT_DATE_TIME']
building_fire['NUMBER_OF_MINUTES'] = building_fire['TIME_DIFFERENCE'].dt.total_seconds()/60
building_fire_minutes = building_fire['NUMBER_OF_MINUTES']
building_fire_minutes
```

```python
# Plot the KDE
plt.figure(figsize=(15, 15))
plt.title("Distribution of number of minutes to arrive on scene", fontsize=25)
plt.xlabel("# of minutes", fontsize=15)
plt.ylabel("Density", fontsize=15)
sns.kdeplot(data=building_fire_minutes, shade=True, clip=(0, 15))
```

```python
# Describe the previous graph
building_fire_minutes.describe()
```

```python
# Print the 3rd quartile
print("The 3rd quartile of the previous graph = {}".format(building_fire_minutes.describe().loc['75%']))
```
